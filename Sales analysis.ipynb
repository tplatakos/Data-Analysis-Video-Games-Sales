```python
import os
import pandas as pd
```

```python
#Read dataframe#
all_data = pd.read_csv("/home/jovyan/demo/vgsales.csv")
all_data.head()
```

```python
# Find NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
df0=all_data.groupby(['Year']).sum()

df0
```

```python
import matplotlib.pyplot as plt

keys = [city for city, df in all_data.groupby(['Platform'])]

plt.bar(keys,all_data.groupby(['Platform']).sum()['NA_Sales'])
plt.ylabel('NA_Sales')
plt.xlabel('Platform')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```
